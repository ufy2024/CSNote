```python
def isValid(s:str):
    left = ["(","[","{"]
    right = {")":"(","]":"[","}":"{"}
    stack = Stack()
    for c in s:
        if c in left:
            stack.push(c)
        elif c in right.keys():
            if stack.top() == right[c]:
                stack.pop()
            else:
                return False
        else:
            #其他字符不做处理
            pass
    if len(stack)==0:
        return True
```

```python
s = "{[([])]}"
isValid(s)
```

```python
def isValid2(s:str):
    left = ["(","[","{"]
    right = {
        ")":"(",
        "]":"[",
        "}":"{"}
    stack = LinkStack()
    for c in s:
        if c in left:
            stack.push(c)
        elif c in right.keys():
            if stack.top() == right[c]:
                stack.pop()
            else:
                return False
        else:
            #其他字符不做处理
            pass
    if len(stack)==0:
        return True
s = "{[([[]])]}"
isValid2(s)
```
